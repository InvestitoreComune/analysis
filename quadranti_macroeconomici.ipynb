```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Crea figura
fig, ax = plt.subplots(figsize=(12, 10))

# Definisci i quadranti
quadrants = {
    'Crescita': {'pos': (0.5, 0.5), 'color': '#4CAF50', 'assets': ['AZIONI +++', 'Obbligazioni +', 'Oro =', 'Cash -']},
    'Inflazione Alta': {'pos': (0.5, -0.5), 'color': '#FF5722', 'assets': ['ORO +++', 'Azioni +', 'Obbligazioni ---', 'Cash ---']},
    'Recessione': {'pos': (-0.5, 0.5), 'color': '#2196F3', 'assets': ['OBBLIGAZIONI LUNGHE +++', 'Oro ++', 'Cash +', 'Azioni ---']},
    'Deflazione': {'pos': (-0.5, -0.5), 'color': '#9C27B0', 'assets': ['OBBLIGAZIONI LUNGHE +++', 'Cash ++', 'Azioni --', 'Oro -']}
}

# Disegna i quadranti
for name, data in quadrants.items():
    x, y = data['pos']
    rect = patches.Rectangle((x-0.45, y-0.45), 0.9, 0.9,
                             linewidth=2, edgecolor='black',
                             facecolor=data['color'], alpha=0.3)
    ax.add_patch(rect)

    # Titolo del quadrante
    ax.text(x, y+0.35, name,
           fontsize=16, fontweight='bold',
           ha='center', va='center')

    # Asset classes
    y_offset = 0.15
    for i, asset in enumerate(data['assets']):
        ax.text(x, y + y_offset - i*0.15, asset,
               fontsize=11, ha='center', va='center',
               bbox=dict(boxstyle='round,pad=0.5', facecolor='white', alpha=0.8))

# Assi - SOLO FINO ALLE FRECCE
ax.plot([0, 0], [-0.95, 0.95], color='black', linewidth=2)  # Asse verticale
ax.plot([-0.95, 0.95], [0, 0], color='black', linewidth=2)  # Asse orizzontale

# Labels degli assi
ax.text(0, 1.15, 'CRESCITA ALTA', fontsize=14, fontweight='bold', ha='center',
       bbox=dict(boxstyle='round,pad=0.5', facecolor='white', edgecolor='black', linewidth=1.5))
ax.text(0, -1.15, 'CRESCITA BASSA', fontsize=14, fontweight='bold', ha='center',
       bbox=dict(boxstyle='round,pad=0.5', facecolor='white', edgecolor='black', linewidth=1.5))
ax.text(-1.15, 0, 'INFLAZIONE\nBASSA', fontsize=14, fontweight='bold',
       ha='center', va='center', rotation=0,
       bbox=dict(boxstyle='round,pad=0.5', facecolor='white', edgecolor='black', linewidth=1.5))
ax.text(1.15, 0, 'INFLAZIONE\nALTA', fontsize=14, fontweight='bold',
       ha='center', va='center', rotation=0,
       bbox=dict(boxstyle='round,pad=0.5', facecolor='white', edgecolor='black', linewidth=1.5))

# Frecce direzionali
ax.annotate('', xy=(0, 0.95), xytext=(0, -0.95),
            arrowprops=dict(arrowstyle='<->', lw=2.5, color='black'))
ax.annotate('', xy=(0.95, 0), xytext=(-0.95, 0),
            arrowprops=dict(arrowstyle='<->', lw=2.5, color='black'))

# Titolo
plt.title('I 4 Quadranti Macroeconomici\nPerformance Asset Classes per Scenario',
         fontsize=18, fontweight='bold', pad=20)

# Impostazioni
ax.set_xlim(-1.6, 1.6)
ax.set_ylim(-1.6, 1.6)
ax.set_aspect('equal')
ax.axis('off')


plt.tight_layout()
plt.show()

print(" Grafico creato!")
```
